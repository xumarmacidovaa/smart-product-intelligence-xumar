# ==============================================================================

# MILESTONE 2: COMPUTER VISION (PRODUCT CATEGORY CLASSIFICATION)

# ==============================================================================

import numpy as np import tensorflow as tf from tensorflow import keras
from tensorflow.keras import layers from sklearn.metrics import
classification_report, confusion_matrix

# 1. Environment and Hyperparameter Configuration

IMAGE_SHAPE = (128, 128, 3) NUM_CLASSES = 5 \# Representing 5 primary
product sub-categories BATCH_SIZE = 32 EPOCHS = 5

print(f”Target Configuration Initialized: Image Dimensions:
{IMAGE_SHAPE}, Total Sub-categories: {NUM_CLASSES}“)

# Simulating image tensor arrays for structural execution logic

# In a real pipeline, these are loaded from the cached image data directory created in Milestone 0

train_images = np.random.rand(100, 128, 128, 3) train_labels =
np.random.randint(0, NUM_CLASSES, size=(100,)) val_images =
np.random.rand(20, 128, 128, 3) val_labels = np.random.randint(0,
NUM_CLASSES, size=(20,)) test_images = np.random.rand(20, 128, 128, 3)
test_labels = np.random.randint(0, NUM_CLASSES, size=(20,))

# ——————————————————————————

# 2. Mandated Baseline Model: Custom CNN from Scratch (Syllabus Requirement)

# ——————————————————————————

print(“— Constructing Baseline Custom CNN Architecture —”) baseline_cnn
= keras.Sequential(\[ layers.Input(shape=IMAGE_SHAPE), layers.Conv2D(32,
(3, 3), activation=‘relu’), layers.MaxPooling2D((2, 2)),
layers.Conv2D(64, (3, 3), activation=‘relu’), layers.MaxPooling2D((2,
2)), layers.Flatten(), layers.Dense(64, activation=‘relu’),
layers.Dense(NUM_CLASSES, activation=‘softmax’)\])

baseline_cnn.compile( optimizer=‘adam’,
loss=‘sparse_categorical_crossentropy’, metrics=\[‘accuracy’\] )

print(“Status: Training Custom CNN Baseline…”) baseline_cnn.fit(
train_images, train_labels, validation_data=(val_images, val_labels),
epochs=EPOCHS, batch_size=BATCH_SIZE, verbose=1 )

# ——————————————————————————

# 3. Advanced Deep Learning Model: MobileNetV2 Transfer Learning

# ——————————————————————————

print(“— Initializing Transfer Learning Pipeline with Pretrained
MobileNetV2 —”)

# Loading the ImageNet pretrained base model, excluding the top classification layers

pretrained_backbone = tf.keras.applications.MobileNetV2(
input_shape=IMAGE_SHAPE, include_top=False, weights=‘imagenet’ )

# Freezing the pretrained backbone weights to prevent gradient updates during training

pretrained_backbone.trainable = False

# Assembling the final deep transfer classification network

transfer_learning_model = keras.Sequential(\[ pretrained_backbone,
layers.GlobalAveragePooling2D(), layers.Dense(128, activation=‘relu’),
layers.Dropout(0.4), \# Implementing regularization against overfitting
layers.Dense(NUM_CLASSES, activation=‘softmax’)\])

transfer_learning_model.compile(
optimizer=keras.optimizers.Adam(learning_rate=0.001),
loss=‘sparse_categorical_crossentropy’, metrics=\[‘accuracy’\] )

print(“Status: Training Deep Transfer Learning Model…”)
transfer_learning_model.fit( train_images, train_labels,
validation_data=(val_images, val_labels), epochs=EPOCHS,
batch_size=BATCH_SIZE, verbose=1 )

# ——————————————————————————

# 4. Comprehensive Class-Balanced Evaluation (Syllabus Requirement)

# ——————————————————————————

print(“— Executing Final Evaluation and Generating Reports —”)
predictions = transfer_learning_model.predict(test_images)
predicted_classes = np.argmax(predictions, axis=1)

# Generating Classification Report to monitor Macro-F1 scores due to rating imbalance

print(“Computer Vision Classification Metrics:”)
print(classification_report(test_labels, predicted_classes,
zero_division=0))

# Generating Confusion Matrix for error analysis mapping

print(“Confusion Matrix Diagnostics:”)
print(confusion_matrix(test_labels, predicted_classes))

print(“: Milestone 2 Computer Vision Pipeline Execution Complete.”)
